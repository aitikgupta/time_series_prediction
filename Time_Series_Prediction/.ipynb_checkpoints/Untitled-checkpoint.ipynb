```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("CMO_MSP_Mandi.csv", delimiter=",")
df2 = pd.read_csv("Monthly_data_cmo.csv", delimiter=",")
```

```python
df=df2
df.describe()
```

```python
df.modal_price = df.modal_price[df.modal_price < 10000]
df.arrivals_in_qtl=df.arrivals_in_qtl[df.arrivals_in_qtl < 300000]
df.modal_price=df.modal_price.fillna(df.modal_price.mean())
```

```python
df.describe()
```

```python
unique_apmc = df.APMC.unique()
unique_commodities = df.Commodity.unique()
```

```python
def plot(time_data, check=None, output_type=None, type_of_data=None):
    if type_of_data == "Commodity" or type_of_data == "APMC" and output_type != None:
        fig = plt.figure(figsize=(20, 10))
        fig.suptitle(check+"--"+output_type, fontsize=20)
        for dfplot in time_data:
            x = pd.to_datetime(dfplot.date).values
            y = dfplot[output_type].values
            yp = y[np.argsort(x)]
            if type_of_data == "Commodity":
                label = dfplot.APMC.values[0]
            else:
                label = dfplot.Commodity.values[0]
            xp = np.array(sorted(x))
            plt.plot(xp, yp, label=label)
        plt.legend()
    else:
        print("Give Correct type of Data: 'APMC' or 'Commodity'")
```

```python
unique_commodity_time_data = []
check_commodity = unique_commodities[34]
commodity_df = df[df.Commodity == check_commodity]
commodity_apmc = commodity_df.APMC.unique()

for apmc in commodity_apmc:
    unique_commodity_time_data.append(commodity_df[commodity_df.APMC == apmc])

plot(unique_commodity_time_data, check_commodity, "arrivals_in_qtl", "Commodity")
plot(unique_commodity_time_data, check_commodity, "modal_price", "Commodity")
```

```python
unique_apmc_time_data = []
check_apmc = unique_apmc[15]
apmc_df = df[df.APMC == check_apmc]
apmc_commodity = apmc_df.Commodity.unique()

for commodity in apmc_commodity:
    unique_apmc_time_data.append(apmc_df[apmc_df.Commodity == commodity])

plot(unique_apmc_time_data, check_apmc, "arrivals_in_qtl","APMC")
plot(unique_apmc_time_data, check_apmc, "modal_price","APMC")
```
