# Monthly Price, Arrival quantity Visualization

### Plot Function

```python
def plot(time_data, title=None, plot_type=None, output_type=None):
    if plot_type == "Commodity" or plot_type == "APMC" or plot_type=="Commodity_msp" or plot_type=="Crop_msp" and output_type!=None:
        fig, ax = plt.subplots(figsize=(10,10))
        fig.suptitle(title+"--"+output_type, fontsize=20)
        if plot_type == "Commodity":
            unique_data = time_data.APMC.unique()
        elif plot_type=="APMC":
            unique_data = time_data.Commodity.unique()
        elif plot_type=="Commodity_msp":
            unique_data=time_data.commodity.unique()
        else:
            unique_data=time_data.Type.unique()
        for data in unique_data:
            df = time_data.get_group(data[0])
            df.plot(y=output_type, ax=ax, label=data[0], legend=False, figsize=(15,10))
        plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=8)
    else:
        print("Invalid Data:\nplot_type: 'Commodity' or 'APMC' or 'Commodity_msp' or 'Crop_msp'\noutput_type: 'modal_price' or 'arrivals_in_qtl' or 'msprice'")
```

```python
df = df2
df.describe()
```

```python
# df.modal_price = df.modal_price[df.modal_price < 10000]
# df.arrivals_in_qtl = df.arrivals_in_qtl[df.arrivals_in_qtl < 300000]
df.modal_price = df.modal_price.fillna(df.modal_price.mean())
```

```python
df.head()
```

```python
unique_apmc = df.APMC.unique()
unique_commodities = df.Commodity.unique()
unique_district = df.district_name.unique()
```

## Commodity type plot

```python
check_commodity = unique_commodities[34]
commodity_df = df[df.Commodity == check_commodity]
commodity_group = commodity_df.groupby('APMC')

plot(commodity_group, check_commodity, 'Commodity', 'arrivals_in_qtl')
```

## APMC type plot

```python
check_apmc = unique_apmc[0]
apmc_df = df[df.APMC == check_apmc]
apmc_group = apmc_df.groupby('Commodity')

plot(apmc_group, check_apmc, 'APMC', 'modal_price')
```
